### Функция для отображения средних спектров выбранных участков

```python
def spectr_grapsh_mean10(cube_1, cube_2, cube_3, cube_4, cube_5, cube_6, cube_7, cube_8, cube_9, cube_10, w_data, n2, n3, n4, n5, n6, n7, n8, n9, n10):

    name_col = list(np.linspace(1,250,250).astype(int).astype(str))
    
    df1 = pd.DataFrame(np.resize(cube_1, (200,250)), columns=name_col)
    df2 = pd.DataFrame(np.resize(cube_2, (200,250)), columns=name_col)
    df3 = pd.DataFrame(np.resize(cube_3, (200,250)), columns=name_col)
    df4 = pd.DataFrame(np.resize(cube_4, (200,250)), columns=name_col)
    df5 = pd.DataFrame(np.resize(cube_5, (200,250)), columns=name_col)
    df6 = pd.DataFrame(np.resize(cube_6, (200,250)), columns=name_col)
    df7 = pd.DataFrame(np.resize(cube_7, (200,250)), columns=name_col)
    df8 = pd.DataFrame(np.resize(cube_8, (200,250)), columns=name_col)
    df9 = pd.DataFrame(np.resize(cube_9, (200,250)), columns=name_col)
    df10 = pd.DataFrame(np.resize(cube_10, (200,250)), columns=name_col)

    df1_mean, df1_std = df1.mean(), df1.std()
    df2_mean, df2_std = df2.mean(), df2.std()
    df3_mean, df3_std = df3.mean(), df3.std()
    df4_mean, df4_std = df4.mean(), df4.std()
    df5_mean, df5_std = df5.mean(), df5.std()
    df6_mean, df6_std = df6.mean(), df6.std()
    df7_mean, df7_std = df7.mean(), df7.std()
    df8_mean, df8_std = df8.mean(), df8.std()
    df9_mean, df9_std = df9.mean(), df9.std()
    df10_mean, df10_std = df10.mean(), df10.std()

    fig = plt.figure(figsize = (35,10))
    ax = fig.add_axes([0,0,1,1])

    plt.grid()
    ax.plot(w_data, df1_mean)
    ax.fill_between(w_data,df1_mean-df1_std,df1_mean+df1_std,alpha=.1,  label='cube №1 area of interest') #lots of algae
    
    plt.xticks(rotation = 90)
    ax.plot(w_data, df2_mean*n2)
    ax.fill_between(w_data, df2_mean*n2-df2_std, df2_mean*n2+df2_std, alpha=.1, label='cube №2 area of interest') #little algae
    
    ax.plot(w_data, df3_mean*n3)
    ax.fill_between(w_data,df3_mean*n3-df3_std,df3_mean*n3+df3_std,alpha=.1,  label='cube №3 area of interest') #lots of algae

    ax.plot(w_data, df4_mean*n4)
    ax.fill_between(w_data,df4_mean*n4-df4_std,df4_mean*n4+df4_std,alpha=.1,  label='cube №4 area of interest') #lots of algae

    ax.plot(w_data, df5_mean*n5)
    ax.fill_between(w_data,df5_mean*n5-df5_std,df5_mean*n5+df5_std,alpha=.1,  label='cube №5 area of interest') #lots of algae

    ax.plot(w_data, df6_mean*n6)
    ax.fill_between(w_data,df6_mean*n6-df6_std,df6_mean*n6+df6_std,alpha=.1,  label='cube №6 area of interest') #lots of algae

    ax.plot(w_data, df7_mean*n7)
    ax.fill_between(w_data,df7_mean*n7-df7_std,df7_mean*n7+df7_std,alpha=.1,  label='cube №7 area of interest') #lots of algae

    ax.plot(w_data, df8_mean*n8)
    ax.fill_between(w_data,df8_mean*n7-df8_std,df8_mean*n7+df8_std,alpha=.1,  label='cube №8 area of interest') #lots of algae

    ax.plot(w_data, df9_mean*n9)
    ax.fill_between(w_data,df9_mean*n7-df9_std,df9_mean*n7+df9_std,alpha=.1,  label='cube №9 area of interest') #lots of algae

    ax.plot(w_data, df10_mean*n10)
    ax.fill_between(w_data,df10_mean*n7-df10_std,df10_mean*n7+df10_std,alpha=.1,  label='cube №10 area of interest') #lots of algae
    
    plt.legend(fontsize=24)

    means_10place = [df1_mean, df2_mean, df3_mean, df4_mean, df5_mean, df6_mean, df7_mean, df8_mean, df9_mean, df10_mean]

    return means_10place
```
